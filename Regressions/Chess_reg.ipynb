```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

```python
chess = pd.read_csv("data/games.csv")
chess.head()
```

```python
chess["white_wins"] = chess["winner"] == "white"
chess["white_wins"]
```

----------------------------------------------

### Logistic Regression

The following will try to predict the winner of a chess game

```python
m = LogisticRegression(max_iter=200)
```

```python
y = chess.white_wins
```

```python
X = pd.get_dummies(chess[["opening_name", "white_rating", "black_rating", "opening_ply"]], drop_first=True)
X.head()
```

```python
_ = m.fit(X, y)
```

```python
m.coef_
```

```python
yhat = m.predict(X)
yhat
```

```python
np.mean(yhat == y)
```
